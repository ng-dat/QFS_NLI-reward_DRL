```python
import json
import math
```

# Train Data

```python
train_set = []
with open(f"./data/train.jsonl") as f:
    for line_count, line in enumerate(f):
        train_set.append(json.loads(line))

train_meeting = []
with open(f"./data/train-meetings.jsonl") as f:
    for line_count, line in enumerate(f):
        train_meeting.append(json.loads(line))
```

```python
print('='*50, 'TRAIN SET', '='*50)
N_train = len(train_set)
print('Number of data:', N_train)

print('-'*111)
avg_query_len = sum([len(line['query'].split()) for line in train_set]) // N_train
print('Average query length:', avg_query_len)

print('-'*111)
avg_ans_len = sum([len(line['answer'].split()) for line in train_set]) // N_train
print('Average answer length:', avg_ans_len)

print('-'*111)
print('Sample of a datapoint:')
data_point = 10
for x in train_set[data_point]:
    print('\t\"{}\" : \"{}\" ,'.format(x, train_set[data_point][x]))
```

```python
print('='*47, 'TRAIN MEETINGS', '='*47)
N_meetings = len(train_meeting)
print('Number of meetings:', N_meetings)

print('-'*111)
avg_len = sum([len((' '.join(line['meeting_transcripts'])).split()) for line in train_meeting]) // N_meetings
print('Average length of meeting transcripts:', avg_len)

print('-'*111)
print('Sample of a meeting data:')
meeting_idx = 8
for x in train_meeting[meeting_idx]:
    print('\t\"{}\" : \"{}\" ,'.format(x, train_meeting[meeting_idx][x]))

```

# Val & Test Data

```python
val_set = []
with open(f"./data/val.jsonl") as f:
    for line_count, line in enumerate(f):
        val_set.append(json.loads(line))

val_meeting = []
with open(f"./data/val-meetings.jsonl") as f:
    for line_count, line in enumerate(f):
        val_meeting.append(json.loads(line))
        
test_set = []
with open(f"./data/test.jsonl") as f:
    for line_count, line in enumerate(f):
        test_set.append(json.loads(line))

test_meeting = []
with open(f"./data/test-meetings.jsonl") as f:
    for line_count, line in enumerate(f):
        test_meeting.append(json.loads(line))
```

```python
print('='*51, 'VAL SET', '='*51)
N_val = len(val_set)
print('Number of data:', N_val)
print('-'*111)
avg_query_len = sum([len(line['query'].split()) for line in val_set]) // N_val
print('Average query length:', avg_query_len)
print('-'*111)
avg_ans_len = sum([len(line['answer'].split()) for line in val_set]) // N_val
print('Average answer length:', avg_ans_len)
print('-'*111)
N_meetings = len(val_meeting)
print('Number of meetings:', N_meetings)
print('-'*111)
avg_len = sum([len((' '.join(line['meeting_transcripts'])).split()) for line in val_meeting]) // N_meetings
print('Average length of meeting transcripts:', avg_len)


print('\n')
print('='*51, 'TEST SET', '='*51)
N_test = len(test_set)
print('Number of data:', N_test)
print('-'*111)
avg_query_len = sum([len(line['query'].split()) for line in test_set]) // N_test
print('Average query length:', avg_query_len)
print('-'*111)
avg_ans_len = sum([len(line['answer'].split()) for line in test_set]) // N_test
print('Average answer length:', avg_ans_len)
print('-'*111)
N_meetings = len(test_meeting)
print('Number of meetings:', N_meetings)
print('-'*111)
avg_len = sum([len((' '.join(line['meeting_transcripts'])).split()) for line in test_meeting]) // N_meetings
print('Average length of meeting transcripts:', avg_len)
```
